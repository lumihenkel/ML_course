## 2 Cost functions a)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define example value pairs
true_values = np.array([1, 2, 3, 4, 5])
predicted_values = np.array([1.1, 1.9, 3.2, 3.8, 5.1])

def cost(true, pred, epsilon=1e-8):
    return (true - pred) ** 2 / (true**2 + epsilon) 

# Calculate normalized squared error
normalized_se = cost(true_values, predicted_values)

# Plot normalized squared error
plt.figure(figsize=(10, 6))
plt.plot(range(len(normalized_se)), normalized_se, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Sample Index')
plt.ylabel('Normalized Squared Error')
plt.title('Normalized Squared Error for Example Value Pairs')
plt.grid(True, alpha=0.3)
plt.show()

print(f"True values: {true_values}")
print(f"Predicted values: {predicted_values}")
print(f"Normalized squared errors: {normalized_se}")
```

## 2 Cost functions d)



```python
def cost_ref(f, y):
    return (np.log10(f + 1) - np.log10(y + 1))**2

true_values = np.array([1, 1])
predicted_values = np.array([10, 100])


plt.figure(figsize=(10, 6))
plt.plot(range(len(true_values)), cost_ref(true_values, predicted_values), 'bo-', linewidth=2, markersize=8)
plt.plot(range(len(true_values)), cost(true_values, predicted_values, epsilon=1), 'ro-', linewidth=2, markersize=8)
plt.xlabel('Sample Index')
plt.ylabel('Cost Function Value')
plt.title('Cost Function: (log(f+1) - log(y+1))')
plt.grid(True, alpha=0.3)
plt.show()

print(cost(true_values, predicted_values, epsilon=1))
print(cost_ref(true_values, predicted_values))
```
